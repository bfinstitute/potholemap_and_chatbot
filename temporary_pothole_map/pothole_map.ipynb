```python
import pandas as pd
import numpy as np
import geopandas as gpd
import folium
```

```python
my_subset = gpd.read_file("potholes_data.gpkg") # potholes_data.gpkg contains a cleaned and reduced version of the 'cleaned' potholes
```

```python
def plot_from_df(df, folium_map):
    for i, row in df.iterrows():
        folium.Marker(
            location=[row.Latitude, row.Longitude],
            tooltip=f"Location {i}",  # optional: you can use a row index or placeholder
            icon=folium.Icon(color='red', icon='info-sign')  # default icon
        ).add_to(folium_map)
    return folium_map
```

```python
small_df = my_subset.head(10) # we will limit to only ten potholes. We have close to 70k records, so attempting to display all would kill the notebook
```

```python
# 1. Set map center and zoom
center = [29.358488, -98.626591]
zoom_start = 10
```

```python
m = plot_from_df(small_df, folium.Map(location=center, zoom_start=zoom_start))
```
